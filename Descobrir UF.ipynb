```python
import pandas as pd
#IBGE
df1 = pd.read_csv('Municípios_IBGE.csv', encoding='latin1', delimiter = ';')
#Cidades
df2 = pd.read_csv('Cidades teste.csv', encoding='latin1', delimiter = ';')

df2
```

```python
s1 = pd.concat([df1['Município']])
s2 = pd.concat([df2['Cidade']])

# encontrar as palavras em comum usando o método intersection
common_words = set(s1).intersection(set(s2))

# converter o conjunto de palavras em uma lista
#common_words = list(common_words)

# acessar as linhas correspondentes nos dataframes originais usando o método isin
result_df1 = df1[df1['Município'].isin(common_words)]


result_df1
```
